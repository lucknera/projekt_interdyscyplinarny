```python
import pandas as pd
import numpy as np
```

```python
faostat = pd.read_csv('../datasets/faostat_crops_and_livestock.csv')
faostat.head()
```

```python
faostat = faostat.drop(columns=['Domain Code', 'Area Code (M49)', 'Element Code', 'Item Code (CPC)', 'Year Code', 'Note', 'Flag', 'Flag Description'])
faostat = faostat.drop(columns=['Domain'])
faostat.head()
```

```python
print(faostat['Element'].unique())
```

```python
faostat = faostat[faostat['Element'] == 'Yield']
faostat = faostat.drop(columns=['Element', 'Unit'])
faostat = faostat.rename(columns={'Value': 'Yield'})
faostat.head()
```

```python
print(faostat['Area'].unique())
```

```python
olek_features = pd.read_csv('../features/olek-features.csv')
olek_features.head()
```

```python
print(olek_features['Area'].unique())
```

```python
pivot = faostat.pivot_table(index='Area', columns=['Item'], values='Yield', aggfunc='mean').reset_index()
pivot.head()
```

```python
pivot = pivot.dropna(axis=1, thresh=10)
pivot = pivot.fillna(0)
pivot.head()
```

```python
pivot.to_csv('../features/mateusz-features.csv', index=False)
```
